```python
score = [98, 77, 88, 90, 75, 99, 100, 85, 78, 90]
print(f"원본: {score}")

score.reverse()#리스트를 역순으로
print(f"역순: {score}")

score.sort() # 오름차순으로 요소를 정렬, 원본 변경
print(f"오름차순: {score}")

score.sort(reverse=True) #내림차순 정렬
print(f"내림차순: {score}")

country = ["Korea", "japan", "CHINA", "america", "***", "한글"]
country1 = country.copy()

country1.sort() # 문자열 정렬 시 대문자가 더 작은 것으로 평가됨
print(f"문자정렬(대소문자구분): {country1}")
country.sort(key = str.lower)#대소문자 무시
print(f"문자정렬(대소문자무시): {country1}")

score2 = sorted(score) #정렬하여 새로운 리스트 생성
print(f"새로운 리스트: {score2}")

sortCountry = sorted(country)
print('새로운 리스트: {}'.format(sortCountry))
```

```python
score = [90, 77, 88, 90, 75, 99, 100, 85, 78, 90]

best_index = score.index(100) # 특정 요소의 인덱스 반환
print(f"100점은 {best_index + 1}번 학생")

score90= score.count(90) #특정 요소값의 개수 반환
print(f"90점을 받은 학생은 {score90}명입니다.")

print(f"전체 학생수는 {len(score)} 명이다")#리스트의 길이 반환
print(f"최고 점수는 {score.index(max(score))}번 학생으로 {max(score)}점이다")#리스트의 요소 중 최대값 반환
print(f"최저 점수는 {min(score)}점입니다.")#리스트의 요소 중 최소값 반환

print(f"전체 평균은 {sum(score) / len(score)}점 이다")
```

```python
s = {1, 2, 3}
print(s)

s.add(1)
print(s)

s.add(4)
print(s)

s.remove(1)
print(s)

s.update([2, 4, 5, 6])
print(s)

s.clear()
print(s)
```

```python
#union(합집합)
s = {1, 2, 3}
t = {1, 7, 6, 3}
print(s.union(t))
#intersection(교집합)
print(s & t)
#difference(차집합)
print(s-t)
print(t.difference(s))
```

```python
aList = [1,2,3,4,5]
bList = ['하나', '둘', '셋', '넷', '다섯']
dict_digit = dict()
for i in range(len(aList)):
  dict_digit[aList[i]] = bList[i]

print(dict_digit)
```

```python
# zip함수로 묶기
digit2 = dict()
print(list(zip(aList, bList)))
for k, v in zip(aList, bList):
  digit2[k] = v
print(digit2)
```

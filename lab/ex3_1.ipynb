#**반복문에서의 else**
반복문 완전히 종료될 때 실행됨(break 걸릴 때는 실행 안됨)
반복이 완전히 거짓일 때 사용할 수 있음

```python
#소수 구하기
num = int(input('입력:'))
for i in range(2, num):
  if num % i == 0:
    print('{}는 소수가 아님'.format(num))
    break
else:
  print('{}는 소수'.format(num))
```

```python
# 20보다 작은 소수 판별
#num = int(input('입력:'))
for n in range(2, 21):
  for i in range(2, n):
    if n % i == 0:
      print('{}는 소수가 아님'.format(n))
      break
  else:
    print('{}는 소수'.format(n))
```

```python
#약수 구하기
num = 20
for i in range(2, num+1):
  print('{}의 약수는 '.format(i), end = ' ')
  for j in range(1, i+1):
    if i % j == 0:
      print(j, end = ' ')
  print()
```

```python
#10진수를 2진수로
num = int(input('10진수 입력 : '))
str1 = ''
while num > 0:
  r = num % 2
  str1 = str(r) + str1
  num = num // 2
print('2진수 변환은 {}'.format(str1))
```

```python
num = random.randint(1, 100)

for i in range(5):
  answer = int(input("난수 맞추기 입력 : "))
  if answer == num:
    print("정답!")
    break
  elif answer < num:
    print("{}은 난수보다 작습니다.".format(answer))
  else:
    print("{}은 난수보다 큽니다.".format(answer))
else:
  print("실패")

```
